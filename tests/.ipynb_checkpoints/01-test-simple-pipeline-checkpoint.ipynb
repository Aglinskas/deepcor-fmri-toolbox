```python
indir = '../Data/fMRI-Data/studyforrest-fmriprep'
epi_path = os.path.join(indir,'sub-01/ses-localizer/func/sub-01_ses-localizer_task-objectcategories_run-1_bold_space-MNI152NLin2009cAsym_preproc.nii.gz')
confounds_path = os.path.join(indir,'sub-01/ses-localizer/func/sub-01_ses-localizer_task-objectcategories_run-1_bold_confounds.tsv')
gm_mask_path = os.path.join(indir,'mask_roi.nii')
cf_mask_path = os.path.join(indir,'mask_roni.nii')
```

```python
output_dir = './Data/DeepCor-Outputs'
```

```python
assert os.path.exists(epi_path), 'epi_path does not exist'
assert os.path.exists(epi_path), 'epi_path does not exist'
assert os.path.exists(epi_path), 'epi_path does not exist'
assert os.path.exists(epi_path), 'epi_path does not exist'
```

```python
from deepcor import DeepCorDenoiser
denoiser = DeepCorDenoiser(
    model_version='v2',
    latent_dims=(8, 8),
    n_epochs=100
)
```

```python
denoised_data = denoiser.fit_denoise(
    epi_path=epi_path,
    confounds_path=confounds_path,
    gm_mask_path=gm_mask_path,
    cf_mask_path=cf_mask_path,
    output_dir=output_dir,
)
```
